What is the primary goal of data anonymization? The primary goal is to protect individual privacy while maintaining data utility for research and analysis by removing or modifying identifying information

What is the Havasupai Tribe case about? A case where Arizona State University researchers used blood samples from the tribe for genetic studies beyond diabetes (the original consent) without proper informed consent, raising issues of research ethics and tribal sovereignty.

What was the largest health data breach in history? The Anthem breach in 2015, affecting approximately 79 million people's health records, including names, birthdates, Social Security numbers, and health information.

What is the pulse oximeter controversy about? Concerns about racial bias in pulse oximeters, which tend to overestimate oxygen levels in darker-skinned patients, potentially leading to delayed or missed treatment.

What is the role of an Institutional Review Board (IRB) in research? Reviews research proposals
Ensures protection of human subjects
Evaluates ethical considerations
Monitors ongoing research
Ensures compliance with regulations

What is the difference between spreadsheet software and database software for health data management? Spreadsheet:

Simple data organization
Individual use
Limited data volume
Basic analysis
Database:

Complex data relationships
Multi-user access
Large data volumes
Advanced security features

What is a conflict of interest in public health research? A situation where professional judgment regarding a primary interest (research integrity) may be influenced by a secondary interest (financial gain, personal advancement).

What is informed consent in the context of health data collection? A process where participants:

Receive full information about the study
Understand their rights
Voluntarily agree to participate
Know how their data will be used

What is data masking in the context of de-identification of health data?: Technique to create a structurally similar but inauthentic version of data by obscuring/replacing sensitive information while maintaining data format and utility.

What is the principle of k-anonymity in the context of de-identification of health data?: A property ensuring that for any combination of identifying attributes, there are at least k individuals who share those attributes, preventing unique identification.

What is the main concern about conflicts of interest in public health research?: The potential for bias in research design, conduct, or interpretation that could compromise scientific integrity and public trust.

What is the first step in managing conflicts of interest in public health research?: Disclosure of potential conflicts to relevant parties (institutions, journals, funders).

What is the role of independent oversight in managing conflicts of interest in public health research?: Monitor research conduct
Review conflict declarations
Ensure compliance with policies
Recommend management strategies

What is one strategy for mitigating conflicts of interest in public health research?: Implementation of blind analysis or third-party data analysis to reduce potential bias from conflicted researchers.

What is the potential impact of not properly managing conflicts of interest in public health research?: Compromised research integrity
Loss of public trust
Biased results
Potential harm to participants
Damage to institutional reputation
