## Data Visualization

```python
#Use seaborn to create a scatterplot of 'age' vs 'plalelets'
seaborn.scatterplot(x='age', y='platelets', data=df)
```

```python
#Recreate the plot and add the titile 'Age vs Platelets'
#Add the horizontal axis title "Age [years]"
#Add the vertical axis title "Platelets [kiloplatelets/mL]"
seaborn.scatterplot(x='age', y='platelets', data=df)
pyplot.title('Age vs Platelets')
pyplot.xlabel('Age [years]')
pyplot.ylabel('Platelets [kiloplatelets/mL]')
pyplot.show()
```

```python
seaborn.scatterplot(
    data = df,
    x = 'age'
    y = 'platelets'
    
)
```

```python
#Recreate the plot and distinguish the points by the 'death' column
seaborn.scatterplot(
    data = df,
    x = 'age',
    y = 'platelets',
    hue = 'death'
)
```

```python
#Recreate the plot but distinguish the plots in the death column by changing the style of the point
seaborn.scatterplot(
    data = df,
    x = 'age',
    y = 'platelets',
    hue = 'death',
    style = 'death'
)
```

```python
seaborn.scatterplot(data = df, x = 'age', y = 'serum_sodium')
```

```python
#Recreate the plot but distinguish each point the the serum_sodium column by hue
seaborn.scatterplot(data = df, x = 'age', y = 'serum_sodium', hue = 'serum_sodium')
```

```python
#Creat a histogram of plot of the 'age' column and the serum_sodium column
seaborn.histplot(data = df, x = 'age')
seaborn.histplot(data = df, x = 'serum_sodium')
```
