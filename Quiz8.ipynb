```python
import numpy
import pandas
from plotly import express
from plotly import io
```

```python
# Seed the pseudo-random number generator
numpy.random.seed(28)

# Create an array of random values from a normal distribution with a mean of 85 and a standard deviation of 10 and assign it to the variable dBP
dBP = numpy.random.normal(85, 10, 100000)
print(dBP)
# Calculate the population mean
dBP.mean()
#Select a random sample of 40 individuals from the population and assign the created numpy array to the variable sample.
sample = numpy.random.choice(dBP, 40, replace=False)
print(sample)
# Calculate the sample mean
sample.mean()
print(sample.mean())
# Calculate the sample standard deviation
sample_std = sample.std(ddof=1)
print(sample_std)
# Calculate the standard error of the sample mean and assign the value to the variable
standard_error = sample_std / (len(sample) ** 0.5)
print(standard_error)
# Calculate the confidence coefficient for a 95% confidence level, given the sample size and degrees of freedom, and assign the value to the variable
confidence_coefficient = 1.96
# alculate the lower bound of the confidence interval for the sample mean and assign the result to the variable lower_bound
lower_bound = sample.mean() - (confidence_coefficient * standard_error)
print(lower_bound)
# Calculate the upper bound of the confidence interval for the sample mean and assign the result to the variable upper_bound

upper_bound = sample.mean() + (confidence_coefficient * standard_error)
print(upper_bound)
#Create an array of 1000 bootstrapped sample means (using the sample size of 40) and assign the array to the variable bootstrapped_means and sort the values in the array in ascending order.
bootstrapped_means = numpy.array([numpy.random.choice(sample, len(sample), replace=True).mean() for _ in range(1000)])
bootstrapped_means.sort()
print(bootstrapped_means)
print(bootstrapped_means.sort())
# Show the bootstrapped value representing the 2.5th percentile
bootstrapped_means[int(len(bootstrapped_means) * 0.025)]
print(bootstrapped_means[int(len(bootstrapped_means) * 0.025)])
# Show the bootstrapped value representing the 97.5th percentile
bootstrapped_means[int(len(bootstrapped_means) * 0.975)]
print(bootstrapped_means[int(len(bootstrapped_means) * 0.975)])








```
